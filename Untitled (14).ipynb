```python





Q1. What is multithreading in python? why is it used? Name the module used to handle threads in python.




ANSWER :- Multithreading in Python refers to the concurrent execution of multiple threads within a single process. 
A thread is a lightweight sub-process that can perform tasks independently, sharing the same memory space as other
threads within the same process. Multithreading is used to achieve parallelism and improve the efficiency of 
programs by allowing multiple tasks to be executed concurrently.

In Python, multithreading is used for several purposes:

CONCURRENCY : Multithreading enables concurrent execution of multiple tasks, allowing programs to make progress on
multiple tasks at the same time. This is particularly useful for I/O-bound tasks, where threads can perform 
operations like reading from or writing to files, making network requests, or interacting with databases
without blocking the execution of other threads.

RESPONSIVENESS : By using threads, you can keep the main program responsive while executing time-consuming tasks
in the background. For example, you can use threads to perform lengthy computations or handle user interface 
updates concurrently, ensuring that the program remains interactive and doesn't appear to hang or freeze.

RESOURSE SHARING : Threads can share data and resources within the same process. This allows for efficient 
communication and coordination between different parts of a program. However, 
it's important to synchronize access to shared resources to prevent race conditions or data corruption.






----------------------------------------------------------------------------------------------------------------------------------------------------



Q2. Why threading module used? rite the use of the following functions
1. activeCount()
2. currentThread()
3. enumerate()



ANSWER :- The threading module in Python is used for creating and managing threads. It provides a high-level
interface and several functions and classes to handle threads effectively.

activeCount(): This function is used to retrieve the number of currently active Thread objects in the current 
thread's ThreadGroup. It returns an integer representing the number of active threads. The active
threads include both daemon and non-daemon threads. This function can be helpful 
for monitoring the number of active threads in a program.

currentThread(): This function returns the currently executing Thread object. It is typically used to obtain a 
reference to the current thread within a thread's code. By calling currentThread(), you can access properties
and methods associated with the current thread, such as its name, identification number (ident), or other attributes.

enumerate(): The enumerate() function returns a list of all currently active Thread objects. It is useful for
obtaining a list of all threads that are currently running or alive. The returned list includes both daemon
and non-daemon threads. Each thread is represented by a Thread object, which can be used to access
various attributes and methods associated with each thread.





----------------------------------------------------------------------------------------------------------------------------------------------------



Q3. Explain the following functions
1. run()
2. start()
3. join()
4. isAlive()



ANSWER :- run(): The run() method is the entry point for the thread's activity. It is the method that will be
executed when a thread is started. By default, it calls the target function passed during
the thread's initialization. WE can also override this method in a custom Thread subclass 
to define the specific behavior that should be executed when the thread starts.


start(): The start() method is used to start the execution of a thread. It creates a new thread of execution
and calls the run() method internally. Once start() is called, the thread is scheduled by the Python
interpreter to run concurrently with other threads. It's important to note that you should never call run()
directly, as it will execute the target function in the same thread, without any concurrency.


join(): The join() method is used to wait for a thread to complete its execution. When join() is called on a thread,
the calling thread is blocked until the target thread finishes its execution. This allows for synchronization,
ensuring that the main thread or any other thread waits for the completion of a specific thread before proceeding.
Additionally, can provide timeout value to join() to specify the maximum amount of time to wait for the thread to finish.


isAlive(): The isAlive() method is used to check if a thread is currently active or alive. It returns a boolean value 
indicating whether the thread is still running. If the thread has completed its execution or hasn't been started yet, 
isAlive() will return False. This method is often used to check the status of a thread and
take appropriate actions based on its current state.





----------------------------------------------------------------------------------------------------------------------------------------------------




Q4.Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.




ANSWER :-


import threading

def print_squares():
for i in range(1, 6):
print(f"Square of {i} is {i**2}")

def print_cubes():
for i in range(1, 6):
print(f"Cube of {i} is {i**3}")

# Create the first thread for printing squares
thread_squares = threading.Thread(target=print_squares)

# Create the second thread for printing cubes
thread_cubes = threading.Thread(target=print_cubes)

# Start both threads
thread_squares.start()
thread_cubes.start()

# Wait for both threads to finish
thread_squares.join()
thread_cubes.join()

print("Done")






In this program, we define two functions print_squares() and print_cubes(), 
which print the squares and cubes of numbers 1 to 5, respectively.

We create two Thread objects, thread_squares and thread_cubes, and assign 
the corresponding target functions to them.

Then, we start both threads using the start() method, which initiates the
execution of the target functions concurrently.

After starting the threads, we call join() on each thread to ensure
that the main program waits for them to complete before moving forward.

Finally, we print "Done" to indicate that both threads have finished executing.






----------------------------------------------------------------------------------------------------------------------------------------------------




Q5. State advantages and disadvantages of multithreading.




ANSWER :-


Advantages of Multithreading:

    
    
Concurrency and Parallelism: Multithreading allows for concurrent execution of multiple tasks, 
leading to increased efficiency and the potential for parallelism on systems with multiple cores.
It enables tasks to progress simultaneously and utilize available system resources effectively.

Responsiveness and Improved User Experience: By using multithreading, you can keep the user interface responsive
while performing time-consuming tasks in the background. This prevents the program from freezing or becoming 
unresponsive, enhancing the overall user experience.

Resource Sharing: Threads can share memory and resources within the same process, allowing efficient communication
and coordination between different parts of a program. This can lead to better utilization of system resources
and increased program efficiency.

Simplified Program Structure: Multithreading can simplify the design and structure of certain programs by 
dividing complex tasks into smaller, manageable threads. This can improve code organization, readability,
and maintainability.



Disadvantages of Multithreading:

    
Complexity and Difficulty of Debugging: Multithreaded programs can be more challenging to develop and debug compared 
to single-threaded programs. Issues such as race conditions, deadlocks, and thread synchronization errors can occur,
making it harder to identify and resolve bugs.

Synchronization Overhead: When multiple threads access shared resources, careful synchronization is required to 
prevent data corruption or inconsistent states. Implementing proper synchronization mechanisms, such as locks 
or semaphores, can introduce additional overhead and complexity to the program.

Reduced Determinism and Potential for Non-deterministic Behavior: Multithreading can introduce non-deterministic 
behavior due to the unpredictability of thread scheduling and execution. This can make it harder to reproduce 
and debug certain issues, as the order of thread execution may vary.

Increased Memory Usage: Each thread requires additional memory resources for its stack and execution context.
In programs with a large number of threads or heavy thread usage, this can lead to increased memory usage 
and potentially impact overall system performance.







----------------------------------------------------------------------------------------------------------------------------------------------------



Q6. Explain deadlocks and race conditions.



ANSWER :-



Deadlocks:
A deadlock occurs when two or more threads are blocked indefinitely, waiting for each other to release resources 
or perform a certain action. In a deadlock situation, the threads are unable to proceed, resulting in a program
freeze or deadlock state.

Mutual Exclusion: Resources involved in the deadlock can only be used by one thread at a time (e.g., locks).
Hold and Wait: A thread holds a resource while waiting to acquire additional resources.
No Preemption: Resources cannot be forcibly taken away from a thread; they can only be released voluntarily.
Circular Wait: A cycle exists in the resource allocation graph, where each thread is waiting for a resource held by another thread.
To prevent deadlocks, it is crucial to ensure that at least one of the deadlock conditions is not met. 
Strategies such as resource ordering, deadlock detection algorithms, and deadlock avoidance techniques 
can be employed to mitigate the risk of deadlocks.


Race Conditions:
A race condition occurs when the behavior of a program depends on the relative ordering or timing of events, 
which can lead to unpredictable or incorrect results. It arises when multiple threads access shared resources 
or modify shared data concurrently without proper synchronization. Race conditions can result in data corruption,
inconsistent states, or unexpected output.
Race conditions typically occur due to the interleaved execution of threads, where the order of execution 
and access to shared resources are non-deterministic. To prevent race conditions, synchronization mechanisms
like locks, semaphores, or mutexes are employed to ensure exclusive access to shared resources. By properly
synchronizing the access to shared data, race conditions can be avoided.

It's important to note that both deadlocks and race conditions can be complex to detect and debug. 
Proper understanding of concurrent programming principles, careful design, and the use of synchronization
mechanisms are essential to avoid these issues and ensure the correctness and reliability of multithreaded programs.







```
