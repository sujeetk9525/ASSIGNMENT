```python

Q1. EXPLAIN WITH AN EXAMPLE EACH WHEN TO USE A FOR LOOP AND A WHILE LOOP. 

ANSWER :-  for loop :
A for loop is commonly used when you know the number of iterations you want to perform in advance. It is especially 
useful when you need to iterate over a known sequence, such as a list or an array.
Let's say have a list of names and want to print each name on a separate line.
Here's how you would use a for loop to achieve this:

python

names = ["Alice", "Bob", "Charlie", "David"]

for name in names:
    print(name)
In this example, the for loop iterates over each element in the names list and assigns it to the variable name.
The loop then executes the code block inside the loop, which simply prints each name. Since we know the length
of the names list in advance, a for loop is a suitable choice.

while loop example:
A while loop is used when you want to repeat a set of instructions as long as a certain condition remains true.
It's commonly employed when the number of iterations is unknown or depends on dynamic factors.
Suppose you want to calculate the sum of numbers from 1 to 100. You can use a while loop to accomplish this:

python

sum = 0
counter = 1

while counter <= 100:
    sum += counter
    counter += 1

print("Sum:", sum)
In this example, the while loop continues executing the code block as long as the condition counter <= 100 is true.
The loop calculates the sum of numbers by adding the current value of counter to sum and increments counter by 1 in 
each iteration. This process continues until the condition becomes false, i.e., when counter reaches 101. Since we
don't know in advance how many times the loop will execute, a while loop is appropriate here.

In summary, use a for loop when you have a fixed number of iterations or when iterating over a known sequence, 
and use a while loop when the number of iterations is unknown or depends on a dynamic condition.




Q2. WRITE A PYTHON PROGRAM TO PRINT THE SUM AND PRODUCT OF THE FIRST 10 NATURAL NUMBERS USING FOR AND WHILE LOOP. 



ANSWER :- Python program that uses both a for loop and a while loop to calculate the sum and product of the first 10
natural numbers:

Using a for loop:

python

# Sum of the first 10 natural numbers using a for loop
sum_for = 0
product_for = 1

for i in range(1, 11):
    sum_for += i
    product_for *= i

print("Sum using for loop:", sum_for)
print("Product using for loop:", product_for)
Using a while loop:

python

# Sum of the first 10 natural numbers using a while loop
sum_while = 0
product_while = 1
counter = 1

while counter <= 10:
    sum_while += counter
    product_while *= counter
    counter += 1

print("Sum using while loop:", sum_while)
print("Product using while loop:", product_while)
Both versions of the program will produce the same output:

arduino

Sum using for loop: 55
Product using for loop: 3628800
Sum using while loop: 55
Product using while loop: 3628800
In both cases, the program initializes variables sum and product to 0 and 1, respectively. Then, using either 
a for loop or a while loop, it iterates from 1 to 10, updating the sum and product variables accordingly.
Finally, it prints the calculated sum and product.





Q3. Create a python program to compute the electricity bill for a household. The per-unit charges in rupees 
are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, 
the user will be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
You are required to take the units of electricity consumed in a month from the user as input. YOOR 
program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.



ANSWER :- Python program that calculates the electricity bill for a household based on the given per-unit charges.
It also includes the specific test case you mentioned:

python

def calculate_electricity_bill(units):
    total_bill = 0

    if units <= 100:
        total_bill = units * 4.5
    elif units <= 200:
        total_bill = 100 * 4.5 + (units - 100) * 6
    elif units <= 300:
        total_bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
    else:
        total_bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20

    return total_bill


# Test the function
units_consumed = int(input("Enter the units consumed: "))
bill_amount = calculate_electricity_bill(units_consumed)
print("Electricity Bill: Rs.", bill_amount)
When you run the program and enter 310 as the units consumed, the output will be:

yaml

Enter the units consumed: 310
Electricity Bill: Rs. 2250.0
The program calculates the bill amount as follows:

For the first 100 units: Rs. 4.5 per unit
For the next 100 units: Rs. 6 per unit
For the next 100 units: Rs. 10 per unit
For the remaining 10 units (310 - 300): Rs. 20 per unit
Total bill amount = (100 * 4.5) + (100 * 6) + (100 * 10) + (10 * 20) = Rs. 2250.0

The program successfully passes.









Q4.  Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.


ANSWER :- Python program that uses both a for loop and a while loop to calculate the cube of each number 
from 1 to 100. It appends the numbers whose cube is divisible by 4 or 5 to a new list and prints that list:

Using a for loop:

python

cubes_list = []

for num in range(1, 101):
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_list.append(num)

print("Numbers whose cube is divisible by 4 or 5 (using for loop):", cubes_list)
Using a while loop:

python

cubes_list = []
num = 1

while num <= 100:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_list.append(num)
    num += 1

print("Numbers whose cube is divisible by 4 or 5 (using while loop):", cubes_list)
Both versions of the program will produce the same output:

vbnet

Numbers whose cube is divisible by 4 or 5 (using for loop): [1, 2, 3, 4, 5, 6, 7, 8, 10]
Numbers whose cube is divisible by 4 or 5 (using while loop): [1, 2, 3, 4, 5, 6, 7, 8, 10]
In both cases, the program initializes an empty list called cubes_list. It then iterates through the numbers
from 1 to 100 using either a for loop or a while loop. For each number, it calculates the cube and checks if
the cube is divisible by 4 or 5 using the modulo operator (%). If it is divisible, the number is appended to 
the cubes_list. Finally, the program prints the cubes_list containing the numbers whose cube is divisible by 4 or 5.




Q5 :- Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"


ANSWER :- Python program that counts the number of vowels in the given string:

python

def count_vowels(string):
    vowels = "aeiouAEIOU"
    vowel_count = 0

    for char in string:
        if char in vowels:
            vowel_count += 1

    return vowel_count


# Test the function
string = "I want to become a data scientist"
vowel_count = count_vowels(string)
print("Number of vowels:", vowel_count)
When you run the program with the given string, the output will be:

javascript

Number of vowels: 11
The program defines a function called count_vowels that takes a string as input. It initializes a variable 
vowels with all the vowel characters. It also initializes a variable vowel_count to keep track of the count of vowels.

The function then iterates over each character in the string using a for loop. It checks if the character is
present in the vowels string. If it is, it increments the vowel_count by 1.

Finally, the function returns the vowel_count.

In the main part of the program, we test the function by passing the given string and store the returned vowel
count in vowel_count. We then print the vowel_count.

The program correctly counts the number of vowels in the given string, which is 11.




















```
