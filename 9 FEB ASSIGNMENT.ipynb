```python




Q1. Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.



ANSWER :- 

AN example of a Vehicle class with an __init__ method that initializes instance variables
for the name of the vehicle, maximum speed, and average speed:

python

class Vehicle:
def __init__(self, name_of_vehicle, max_speed, average_speed):
self.name_of_vehicle = name_of_vehicle
self.max_speed = max_speed
self.average_speed = average_speed

In this class, the __init__ method takes three parameters: name_of_vehicle, max_speed,
and average_speed. These parameters are used to initialize the respective instance 
variables of the class (self.name_of_vehicle, self.max_speed, and self.average_speed).

WE can create instances of the Vehicle class and access the instance variables as follows:

python

# Create a Vehicle instance
car = Vehicle("Car", 200, 80)

# Access instance variables
print(car.name_of_vehicle)    # Output: Car
print(car.max_speed)          # Output: 200
print(car.average_speed)      # Output: 80
By calling the __init__ method during the creation of a Vehicle instance,
you can provide specific values for the instance variables, 
allowing each instance to have its own set of values for the name,
maximum speed, and average speed.





------------------------------------------------------------------------------------------------------------------------------------------------




Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

ANSWER :- 

Example of a child class Car that inherits from the Vehicle class. It includes a method seating_capacity
which takes a capacity argument and returns the name of the vehicle and its seating capacity:

python

class Car(Vehicle):
def __init__(self, name_of_vehicle, max_speed, average_speed):
super().__init__(name_of_vehicle, max_speed, average_speed)

def seating_capacity(self, capacity):
return f"{self.name_of_vehicle} - Seating Capacity: {capacity} passengers"


In this example, the Car class is defined as a child class of the Vehicle class using class Car(Vehicle). 
The __init__ method is overridden in the Car class to inherit the initialization behavior 
from the parent class using super().__init__(...).

The seating_capacity method is added to the Car class, which takes a capacity argument and returns 
a formatted string containing the name of the vehicle and its seating capacity.

WE can create an instance of the Car class and call the seating_capacity method as follows:

python

# Create a Car instance
car = Car("Car", 200, 80)

# Call the seating_capacity method
capacity = 4
result = car.seating_capacity(capacity)
print(result)   # Output: Car - Seating Capacity: 4 passengers
By creating an instance of the Car class and calling the seating_capacity method, 
you can provide a specific seating capacity value and retrieve the formatted 
string that includes the name of the vehicle and its seating capacity.





------------------------------------------------------------------------------------------------------------------------------------------------




Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.


ANSWER :- 


Multiple inheritance is a feature in object-oriented programming where a class can
inherit attributes and behaviors from more than one parent class.
This means that a child class can inherit from multiple parent classes, 
allowing it to incorporate features from each parent class.





class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} says something...")


class Mammal:
    def walk(self):
        print("Mammal walks...")


class Dog(Animal, Mammal):
    def __init__(self, name, breed):
        Animal.__init__(self, name)
        self.breed = breed

    def bark(self):
        print(f"{self.name} barks!")


# Create an instance of the Dog class
dog = Dog("Buddy", "Labrador")

# Call methods from multiple parent classes
dog.speak()     # Output: Buddy says something...
dog.walk()      # Output: Mammal walks...
dog.bark()      # Output: Buddy barks!





In this example, we have three classes: Animal, Mammal, and Dog. The Animal class 
represents an animal with a name and a speak method. The Mammal class represents 
a mammal with a walk method. The Dog class inherits from both the Animal
and Mammal classes using multiple inheritance.

The Dog class has an __init__ method that calls the __init__ method of both the Animal
and Mammal classes using Animal.__init__(self, name) and Mammal.__init__(self).

By creating an instance of the Dog class, we can access and call methods from both
the Animal and Mammal classes. In the example code, we call the speak method from
the Animal class to make the dog say something, we call the walk method from the 
Mammal class to make the dog walk, and we call the bark method defined in 
the Dog class itself to make the dog bark.

Note: The order of the parent classes mentioned in the class declaration 
(class Dog(Animal, Mammal)) determines the method resolution order (MRO) 
in case of conflicts or method overriding. The MRO defines the order
in which methods are searched and executed.






------------------------------------------------------------------------------------------------------------------------------------------------





Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.



ANSWER :- 



In Python, getter and setter methods are used to provide controlled access to class attributes 
or properties. They are used to retrieve and modify the values of private or protected attributes.
The getter method retrieves the value of an attribute, while the setter method
sets or updates the value of an attribute, allowing for encapsulation and data validation.

THE example of a class with a getter and a setter method:


class Person:
    def __init__(self):
        self._name = None  # Private attribute with "_" prefix convention

    def get_name(self):
        return self._name

    def set_name(self, name):
        if isinstance(name, str):
            self._name = name
        else:
            raise ValueError("Name should be a string.")

# Create an instance of the Person class
person = Person()

# Set the name using the setter method
person.set_name("John")

# Get the name using the getter method
name = person.get_name()
print(name)  # Output: John




In this example, we have a Person class with a private attribute _name. 
The getter method get_name retrieves the value of the _name attribute, 
and the setter method set_name sets or updates the value of the _name attribute.
The setter method includes a simple data validation check to ensure that
the name provided is a string.

By creating an instance of the Person class, we can use the setter method set_name
to set the name attribute to "John". We can then use the getter method get_name 
to retrieve the value of the name attribute, which will be "John". The use of 
getter and setter methods allows us to control access to the attribute and 
enforce any necessary validation or logic.





------------------------------------------------------------------------------------------------------------------------------------------------




Q5. What is method overriding in python? Write a python code to demonstrate method overriding.




ANSWER :- Method overriding is a feature in object-oriented programming that allows a subclass
to provide a different implementation of a method that is already defined in its superclass.
This allows the subclass to inherit the method from the superclass but modify its behavior to suit its specific needs.

EXAMPLE:


class Animal:
    def make_sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def make_sound(self):
        print("Dog barks")

class Cat(Animal):
    def make_sound(self):
        print("Cat meows")

# Create instances of the classes
animal = Animal()
dog = Dog()
cat = Cat()

# Call the overridden method on each instance
animal.make_sound()  # Output: Animal makes a sound
dog.make_sound()     # Output: Dog barks
cat.make_sound()     # Output: Cat meows


In the above example, we have a base class Animal with a method make_sound(). The Dog and Cat classes inherit 
from Animal and override the make_sound() method with their own implementations.

When we create instances of Animal, Dog, and Cat and call the make_sound() method on each instance, 
Overridden method each subclass is invoked, producing different outputs based on specific implementation in each subclass.













```
