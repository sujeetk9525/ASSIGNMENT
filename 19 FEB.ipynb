```python













Q1. What is Flask Framework? What are the advantages of Flask Framework ?




ANSWER :-


Flask is a lightweight web framework for Python that allows developers to build web applications quickly and with simplicity.
It follows the WSGI (Web Server Gateway Interface) standard and is based on the Werkzeug toolkit and the Jinja templating engine.

Advantages of Flask Framework:

1. Simplicity and Minimalism: Flask has a minimalistic design and provides a straightforward API, making it easy to understand
and use It allows developers to start building web applications with minimal boilerplate code and complexity.

2. Flexibility and Extensibility: Flask provides a flexible structure, allowing developers to choose the components they need
for their specific requirements. It follows a modular design, and its functionality can be extended by using various 
Flask extensions or integrating with other Python libraries.

3. Routing and URL Mapping: Flask offers a powerful routing system that maps URLs to specific functions or views. This enables
developers to define the routes and associated logic for handling HTTP requests in a clear and concise manner.

4. Template Engine Support: Flask integrates with the Jinja templating engine, which allows for the creation of dynamic and
reusable HTML templates. Jinja provides a simple syntax and powerful features like template inheritance, macros, and filters.

5. Built-in Development Server: Flask comes with a built-in development server, allowing developers to quickly test and iterate 
on their web applications during the development process. This server is convenient for local development and debugging.

6. Extensive Documentation and Active Community: Flask has comprehensive and well-maintained documentation, providing clear 
explanations and examples. It also has a large and active community of developers, which means there are numerous tutorials,
resources, and third-party packages available for support and integration.

7. Scalability: Flask is lightweight and can be used to build both small and large-scale applications. It allows developers to 
add additional components or integrate with other libraries as needed, providing scalability and adaptability to handle 
growing application requirements.

8. Testing and Debugging: Flask provides an intuitive testing framework that simplifies the testing of web applications. 
It also offers built-in debugging tools, such as the interactive debugger and error handling mechanisms, which 
aid in identifying and resolving issues during development.








---------------------------------------------------------------------------------------------------------------------------------







Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.



ANSWER :-



app = Flask(__name__)

@app.route('/')
def hello():
return 'Hello World!!'

if __name__ == '__main__':
app.run()

.py extension (e.g., app.py). Make sure you have Flask installed (pip install flask) 
before running the code.

To run the application, open a terminal or command prompt, navigate to the directory where WE saved the file
and run the following command:


python app.py
You should see output similar to this:

csharp

 * Running on http://127.0.0.1:5000/ (Press CTRL+C to quit)
Now, open a web browser and go to http://127.0.0.1:5000/ or http://localhost:5000/. WE should see the "Hello World!!"
message displayed in the browser.








---------------------------------------------------------------------------------------------------------------------------------




Q3. What is App routing in Flask? Why do we use app routes?




ANSWER :-


In Flask, app routing refers to the process of mapping URL routes to specific functions or views within a Flask application.
It allows you to define different endpoints (URLs) that correspond to different parts or functionalities of your web application.

App routes are defined using the @app.route() decorator in Flask. The decorator is used to associate a URL route with
a function or view that will handle the request to that route. Here's an example:


app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/contact')
def contact():
    return 'Contact Page'

if __name__ == '__main__':
    app.run()
    
In this example, we have defined three app routes: '/', '/about', and '/contact'. Each route is associated with a 
corresponding function (index(), about(), and contact()) that returns a response to be displayed when that 
specific URL is accessed.

When a user visits the root URL '/', the index() function is called and returns the string 'Home Page'. Similarly, 
when a user visits '/about', the about() function is called and returns 'About Page', and so on.




---------------------------------------------------------------------------------------------------------------------------------




Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.



ANSWER :-




app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def details():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
py extension (e.g., app.py). Make sure you have Flask installed (pip install flask) before running the code.

To run the application, open a terminal or command prompt, navigate to the directory where you saved the file,
and run the following command:




* Running on http://127.0.0.1:5000/ (Press CTRL+C to quit)
Now, open a web browser and go to http://127.0.0.1:5000/welcome to see the "Welcome to ABC Corporation" message. 
To view the details on the root URL, go to http://127.0.0.1:5000/. You should see the company details displayed.




---------------------------------------------------------------------------------------------------------------------------------






Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.




ANSWER :- 

In Flask, the url_for() function is used for URL building. It generates a URL for a specific view or 
endpoint by taking into account the routing rules defined in the application.

The url_for() function accepts the name of the view or endpoint as its first argument and any additional keyword 
arguments that correspond to the dynamic parts of the URL route. It automatically handles URL generation,
considering the route patterns and variables defined in the routes.

Here's an example to demonstrate the usage of the url_for() function:



app = Flask(__name__)

@app.route('/')
def index():
return 'Home Page'

@app.route('/user/<username>')
def user_profile(username):
return f'Profile page of {username}'

@app.route('/product/<int:product_id>')
def product_details(product_id):
return f'Details of product {product_id}'

@app.route('/logout')
def logout():
return 'Logout Page'

if __name__ == '__main__':
with app.test_request_context():
# Generating URLs using url_for()
home_url = url_for('index')
profile_url = url_for('user_profile', username='john')
        product_url = url_for('product_details', product_id=123)
        logout_url = url_for('logout')

        # Printing the generated URLs
        print(f'Home URL: {home_url}')
        print(f'Profile URL: {profile_url}')
        print(f'Product URL: {product_url}')
        print(f'Logout URL: {logout_url}')
In this example, we define several routes with different patterns and variables. Within the if __name__ == '__main__' block, we use the url_for() function to generate URLs for specific views.

By calling url_for() with the name of the view function as the first argument and providing the necessary keyword arguments for any dynamic parts in the URL, we can generate the corresponding URLs. The generated URLs are stored in variables (home_url, profile_url, product_url, logout_url) and printed to the console.

When you run this code, you'll see the generated URLs printed in the console. These URLs are dynamically generated based on the defined routes and provided arguments.

Note: It's important to have the application context when using url_for(), as shown in the example using with app.test_request_context().





```
